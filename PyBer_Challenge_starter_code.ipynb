# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_per_city_type_df = pyber_data_df.groupby(['type']).count()['ride_id']
rides_per_city_type_df.head()
```

```python
# 2. Get the total drivers for each city type
drivers_per_city_type_df = city_data_df.groupby(['type']).sum()['driver_count']
drivers_per_city_type_df
```

```python
#  3. Get the total amount of fares for each city type
fares_per_city_type_df = pyber_data_df.groupby(['type']).sum()['fare']
fares_per_city_type_df
```

```python
#  4. Get the average fare per ride for each city type. 
average_fares_per_ride_df = pyber_data_df.groupby(['type']).mean()['fare']
average_fares_per_ride_df
```

```python
# 5. Get the average fare per driver for each city type. 
pyber_avg_driver = pyber_data_df.groupby(["type"]).sum()["fare"] / city_data_df.groupby(["type"]).sum()["driver_count"]
pyber_avg_driver

#city_data_df.head()
#pyber_data_df.head()
```

```python
#  6. Create a PyBer summary DataFrame. 
#pyber_summary_df=pd.concat([rides_per_city_type_df, drivers_per_city_type_df, fares_per_city_type_df,average_fares_per_ride_df,pyber_avg_driver],axis=1)
#pyber_summary_df.set_axis(['Total Rides', 'Total Drivers', 'Total Fares','Average Fare per Ride','Average Fare per Driver'], axis='columns', inplace=True)
#type(pyber_summary_df)
frame = {'Total Rides' :rides_per_city_type_df,
         'Total Drivers' : drivers_per_city_type_df, 
         'Total Fares':fares_per_city_type_df, 
         'Average Fare per Ride':average_fares_per_ride_df, 
         'Average Fare per Driver':pyber_avg_driver}
pyber_summary_df = pd.DataFrame(frame)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name

pyber_summary_df.index.name = None

pyber_summary_df
```

```python
#  8. Format the columns.
#pd.options.display.float_format =  '${:,.2f}'.format
pyber_summary_df.loc[:, 'Total Rides'] =pyber_summary_df['Total Rides'].map('{:,.0f}'.format)
pyber_summary_df.loc[:, 'Total Drivers'] =pyber_summary_df['Total Drivers'].map('{:,.0f}'.format)
pyber_summary_df.loc[:, 'Total Fares'] =pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df.loc[:, 'Average Fare per Ride'] =pyber_summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
pyber_summary_df.loc[:, 'Average Fare per Driver'] =pyber_summary_df['Average Fare per Driver'].map('${:,.2f}'.format)

pyber_summary_df


```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
new_data_frame = pyber_data_df.groupby(['type','date']).sum()['fare']
new_data_frame.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
new_data_frame = pyber_data_df.groupby(['type','date']).sum()['fare']
new_data_frame.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
new_data_frame = new_data_frame.reset_index()
new_data_frame.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
new_data_frame_pivot = new_data_frame.pivot(index = 'date', columns = 'type', values = 'fare')
new_data_frame_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_data_frame_pivot_loc = new_data_frame_pivot.loc['2019-01-01':'2019-04-29']
new_data_frame_pivot_loc.head()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
new_data_frame_pivot_loc.index = pd.to_datetime(new_data_frame_pivot_loc.index)
new_data_frame_pivot_loc
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_data_frame_pivot_loc.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
week_df = new_data_frame_pivot_loc.resample("W").sum()
week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 


week_df.plot(figsize = (25,10))


plt.title("Total Fare by City Type", fontsize = 25)
plt.ylabel("Fare ($)", fontsize = 25)
plt.xlabel("Date", fontsize = 20).set_visible(False)


#Best place for labels
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="center", title="Type")
lgnd.get_title().set_fontsize(12)


# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


plt.savefig("../Resources/Fig10.png")

plt.show()
```
